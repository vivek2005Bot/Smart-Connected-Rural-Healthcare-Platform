```python
class Recommendation:
    def __init__(self,input_symtoms):
        if "," in input_symtoms:
            self.input_symtoms=input_symtoms.split(",")
        else:
            self.input_symtoms=input_symtoms.split(" ")

        path="C:\\Users\\ASUS\\Downloads\\archive\\Backend"
        os.chdir(path)
        self.model=pickle.load(open("model.pkl","rb"))
        self.le=pickle.load(open("encoder.pkl","rb"))  
        self.columns=pickle.load(open("col.pkl","rb"))
        
        self.dataset_diets=pd.read_csv("data sets\\diets.csv")
        self.dataset_description=pd.read_csv("data sets\\description.csv")
        self.dataset_medication=pd.read_csv("data sets\\medications.csv")
        self.dataset_precautions=pd.read_csv("data sets\\precautions_df.csv")
        self.dataset_workout=pd.read_csv("data sets\\workout_df.csv")
        self.dataset_precautions=self.dataset_precautions.drop(columns=["Unnamed: 0"],axis=1)
        self.dataset_workout=self.dataset_workout.iloc[:,2:]


        self.disease=self.predict()
        
    
    def predict(self):
        input_vector=np.zeros(len(self.columns))
        for syms in self.input_symtoms:
            syms_name=difflib.get_close_matches(syms,self.columns)[0]
            index=np.where(self.columns==syms_name)[0][0]
            input_vector[index]=1

        input_vector=np.reshape(input_vector,newshape=(1,-1))
        return self.le.inverse_transform(self.model.predict(input_vector))[0]
    
    def get_diet(self):
        if self.disease not in self.dataset_diets["Disease"].values:
            return ["No diet info found"]

        return self.dataset_diets[self.dataset_diets["Disease"]==self.disease].iloc[:,1:].values[0][0]
    
    def get_description(self):
        if self.disease not in self.dataset_description["Disease"].values:
            return ["No diet info found"]

        return self.dataset_description[self.dataset_description["Disease"]==self.disease].iloc[:,1:].values[0][0]
    
    def get_medication(self):
        if self.disease not in self.dataset_medication["Disease"].values:
            return ["No diet info found"]

        return self.dataset_medication[self.dataset_medication["Disease"]==self.disease].iloc[:,1:].values[0][0]
    
    def get_precautions(self):
        if self.disease not in self.dataset_precautions["Disease"].values:
            return ["No diet info found"]

        return self.dataset_precautions[self.dataset_precautions["Disease"]==self.disease].iloc[:,1:].values[0]
    
    def get_workout(self):
        if self.disease not in self.dataset_workout["disease"].values:
            return ["No diet info found"]

        return self.dataset_workout[self.dataset_workout["disease"]==self.disease].iloc[:,1:].values.tolist()
```
